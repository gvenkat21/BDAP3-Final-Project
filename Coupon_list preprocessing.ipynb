## Import the Coupon list CSV files into Python

```python
import pandas as pd
```

```python
coupon_list_test = pd.read_csv("C:\Users\gvenk\Desktop\data\Coupon Purchase Prediction\coupon_list_test.csv")
coupon_list_test.head(5)
```

```python
coupon_list_train = pd.read_csv("C:\Users\gvenk\Desktop\data\Coupon Purchase Prediction\coupon_list_train.csv")
coupon_list_train.head(5)
```

## Merge the train and test file for data preprocessing

```python
## Mark coupon_list_train file as 'Train' so that we can identify and separate train and test file later

coupon_list_train['Mark'] = 'Train'
coupon_list_train.head(5)
```

```python
## Mark coupon_list_test file as 'Test' so that we can identify and separate train and test file later

coupon_list_test['Mark'] = 'Test'
coupon_list_test.head(5)
```

```python
## Convert all headers to lower case

coupon_list_train.columns = map(str.lower, coupon_list_train.columns)
coupon_list_test.columns = map(str.lower, coupon_list_test.columns)
coupon_list_train.head(5)
```

```python
## Merge train and test
frames = [coupon_list_train, coupon_list_test]
coupon_list = pd.concat(frames)
coupon_list.tail(5)
```

## Perform some basic data cleaning

### Convert Japanese text to English

```python
## Import the translation tables

coupon_text_trans = pd.read_excel("C:\Users\gvenk\Desktop\data\Coupon Purchase Prediction\coupon_text_translation.xlsx")
print(coupon_text_trans)
genre_name_trans = pd.read_excel("C:\Users\gvenk\Desktop\data\Coupon Purchase Prediction\genre_name_translation.xlsx")
# print(genre_name_trans)
```

```python
coupon_text_trans.set_index("jp", drop = True, inplace = True)
genre_name_trans.set_index("jp", drop = True, inplace = True)
```

```python
coupon_text_trans_dict = coupon_text_trans.to_dict()
coupon_text_trans_dict = coupon_text_trans_dict['eng']


```

```python
genre_name_trans_dict = genre_name_trans.to_dict()
genre_name_trans_dict = genre_name_trans_dict['eng']
genre_name_trans_dict
```

```python
tmp = []
for word in coupon_list['capsule_text']:
    for (k,v) in coupon_text_trans_dict.items():
        if unicode(word,"utf-8") == k:
            tmp.append(v)

coupon_list['capsule_text'] = tmp
```

```python
tmp = []
for word in coupon_list['genre_name']:
    for (k,v) in genre_name_trans_dict.items():
        if unicode(word,"utf-8") == k:
            tmp.append(v)

coupon_list['genre_name'] = tmp
```
